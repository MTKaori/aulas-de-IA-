```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

# Carregar os dados
iris = load_iris()
```

```python
# --- PONTO CRUCIAL ---
# Para Aprendizado Não Supervisionado, usamos APENAS os dados (X).
# Não usamos os rótulos (y) para treinar.
X = iris.data
y = iris.target # Usaremos o 'y' apenas NO FINAL, para comparar

print(f"Formato dos dados (X): {X.shape}")
print("NÃO usaremos os rótulos (y) para treinar, apenas para validar no final.")
```

```python
# Lista para armazenar a inércia (WCSS) para cada K
wcss = []
K_range = range(1, 11) # Testando K de 1 a 10

for k in K_range:
    kmeans = KMeans(n_clusters=k, init='k-means++', n_init=10, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_) # 'inertia_' é o valor WCSS
```

```python
# Plotar o gráfico do Método do Cotovelo
plt.figure(figsize=(8, 5))
plt.plot(K_range, wcss, marker='o')
plt.title('Método do Cotovelo (Elbow Method)')
plt.xlabel('Número de Clusters (K)')
plt.ylabel('Inércia (WCSS)')
plt.xticks(K_range)
plt.grid(True)
plt.show()
```

Análise: Você verá que o "cotovelo" (a maior "dobra" no gráfico) acontece exatamente no K=3. Isso significa que o algoritmo, sem saber de nada, sugere que os dados se agrupam naturalmente em 3 clusters!

```python
# Definimos o número de clusters como 3
k = 3

# Criar e treinar o modelo
# n_init=10 executa o algoritmo 10x com centroides diferentes e pega o melhor
kmeans_model = KMeans(n_clusters=k, init='k-means++', n_init=10, random_state=42)
```

```python
# Treinar e prever os clusters no próprio X
y_kmeans_pred = kmeans_model.fit_predict(X)

print("Modelo treinado!")
print("Clusters encontrados para os 5 primeiros dados:")
print(y_kmeans_pred[:5])
```

```python
# Ver os centros dos clusters (centroides)
print("\nCentroides encontrados:")
print(kmeans_model.cluster_centers_)
```

Esta é a parte mais importante. Vamos comparar os grupos reais (os rótulos y que escondemos) com os grupos que o K-Means encontrou (y_kmeans_pred).

Como os dados têm 4 dimensões (features), vamos plotar um gráfico 2D usando as duas primeiras features (Comprimento e Largura da Sépala).

```python
plt.figure(figsize=(14, 6))

```

```python
# --- Gráfico 1: Os Grupos Reais (Nossa "Resposta") ---
plt.subplot(1, 2, 1)
# Usamos 'y' (os rótulos reais) para colorir
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', s=50, alpha=0.8)
plt.title('Grupos Reais (Espécies de Iris)')
plt.xlabel('Comprimento da Sépala (cm)')
plt.ylabel('Largura da Sépala (cm)')
```

```python
# --- Gráfico 2: Os Grupos Encontrados pelo K-Means ---
plt.subplot(1, 2, 2)
# Usamos 'y_kmeans_pred' (as previsões) para colorir
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans_pred, cmap='viridis', s=50, alpha=0.8)
# Plotar os centroides
centers = kmeans_model.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label='Centroides')
plt.title('Grupos Encontrados (K-Means Clustering)')
plt.xlabel('Comprimento da Sépala (cm)')
plt.legend()

plt.show()
```

```python
# Criar uma nova "flor misteriosa" com medidas [Comp. Sépala, Larg. Sépala, Comp. Pétala, Larg. Pétala]
flor_misteriosa = np.array([[5.1, 3.5, 1.4, 0.2]]) # Medidas de uma Setosa
flor_misteriosa_2 = np.array([[6.7, 3.0, 5.2, 2.3]]) # Medidas de uma Virginica
```

```python
# Usar o modelo para "prever" a qual cluster ela pertence
cluster_previsto_1 = kmeans_model.predict(flor_misteriosa)
cluster_previsto_2 = kmeans_model.predict(flor_misteriosa_2)

print(f"A flor 1 pertence ao cluster: {cluster_previsto_1[0]}")
print(f"A flor 2 pertence ao cluster: {cluster_previsto_2[0]}")
```
